```python
%matplotlib inline

import os
import numpy as np
from numpy.random import normal, uniform
import collections
import pandas as pd
from astropy.table import Table

import triangle
import sncosmo
import simulate_lsst as sl
import lightcurve_analysis as lca
```

```python
data = sncosmo.read_lc('sn.dat')
```

```python
model = sncosmo.Model(source='salt2-extended')
params = {'t0': data.meta['SIM_PEAKMJD'], 'x0': data.meta['SIM_SALT2x0'],
          'x1': data.meta['SIM_SALT2x1'], 'c': data.meta['SIM_SALT2c'],
          'z': data.meta['SIM_REDSHIFT_HELIO']}
model.set(**params)
vparams = ['t0', 'x0', 'x1', 'c']
```

The class has now been renamed <code>lightcurve_analysis</code> (imported as <code>lca</code> for compactness).  The class object(?) is now called <code>LCA</code>.

```python
import math
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

mean = example.fit_res.parameters[1]
variance = example.fit_res.covariance[0][0]
sigma = math.sqrt(variance)
min = example.fit_res.parameters[1] - 8*sigma
max = example.fit_res.parameters[1] + 8*sigma
x = np.linspace(min, max, 100)
```

```python
mean, min, max
```

```python
plt.hist(example.mcmc_res.samples[:,0], normed=True)
plt.plot(x,mlab.normpdf(x,mean,1))
```
